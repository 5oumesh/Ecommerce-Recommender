```python
import pandas as pd

df = pd.read_csv("../data/interim/filtered_sample_final.csv")

print("Rows:", len(df))
print("Users:", df["user_id"].nunique())
print("Products:", df["product_id"].nunique())
```

```python
users = df[["user_id"]].drop_duplicates()

users.to_csv(
    "../data/processed/users.csv",
    index=False
)

print("users.csv saved:", users.shape)
```

```python
products = (
    df[["product_id", "category_code"]]
    .drop_duplicates()
)

products.to_csv(
    "../data/processed/products.csv",
    index=False
)

print("products.csv saved:", products.shape)
```

```python
interactions = df[
    ["user_id", "product_id", "event_type", "event_strength", "event_time"]
]

interactions.to_csv(
    "../data/processed/interactions.csv",
    index=False
)

print("interactions.csv saved:", interactions.shape)
```

```python
import pandas as pd
import os

base_path = "../data/processed"

files = ["users.csv", "products.csv", "interactions.csv"]

for f in files:
    path = os.path.join(base_path, f)
    print(f"\nChecking {f}...")
    if os.path.exists(path):
        df = pd.read_csv(path)
        print("  ✅ Exists")
        print("  Shape:", df.shape)
        print("  Columns:", list(df.columns))
    else:
        print("  ❌ Missing")

```
